###**Christmas Movies (2017-22) Analysis**

This project works on analysis of various Christmas Movies which were launched from 2017-22. It analyses the data by :   


> 1. Finding out which actors appeared the most and the least in christmas movies as a pair that were released from 2017-22.
> 2. Finding out the number of movies done by each of the main cast actor for Christmas Movies released from 2017-22.
> 3. Finding out the most and least voted Christmas Movies from the ones that were released from 2017-2022.
> 4. Finding out the most and least used genre for Christmas Movies released from 2017-2022.
> 5. Finding out the longest and shortest christmas movie released from 2017-22.
> 6. Finding all the Christmas Movies that start with C, have a rating 6.0 + and are of Drama genre.
> 7. Finding out the director who has worked most and least on Christmas Movies released from 2017-22.
> 8. Finding out the per category/certificate count for the Christmas Movies released from 2017-2022.
> 9. Finding out the exactly 7.1 rated Christmas Movies which were released from 2017-2022.
> 10. Finding out the Christmas Movies that have 'cookie' in their description.
> 11. Finding out the most common cast (i.e., the 4 people who have appeared together the most) for the Christmas Movies which were released from 2017-22.






Dataset used : [IMDB Dataset of All Christmas Movies 2016-2022](https://www.kaggle.com/datasets/fc32844f7c01a95f32aac01a62ec15cc73f000a8809a945ee2e113cd4107a54a)


```python
#importing the pandas library 
import pandas as pd
df = pd.read_csv("/content/christmas_movies.csv", encoding = 'unicode_escape', engine ='python')
```

```python
#converting into .xlsx file type and printing the first 20 rows
df.to_excel("christmas_movie_excel.xlsx" , index = False)
df.head(20)
```

```python
#dropping the column 'score' as very few movies have the score and hence will tamper the data 
del df["Score"]
```

```python
#creating a new column of all actors that has cummulative results from cast1,2,3 and 4 separated by commas
df["all_actors"] = df[['cast1', 'cast2', 'cast3', 'cast4']].astype(str).agg(' , '.join, axis=1)
df.head()
```

## Finding out which actors appeared the most and the least in Christmas Movies as a pair that were released from 2017-22.

```python
#10 most common pairs appeared in the christmas movies from 2017-2022
from itertools import combinations
from collections import Counter

count = Counter()
for row in df["all_actors"]:
  row_list = row.split(',') #Creating the sublist and segrating by each comma
  count.update(Counter(combinations(row_list,2)))
count.most_common(10) 
```

```python
#10 most uncommon pairs appeared in the christmas movies from 2017-2022
from itertools import combinations
from collections import Counter

count = Counter()
for row in df["all_actors"]:
  row_list = row.split(',') #Creating the sublist and segrating by each comma
  count.update(Counter(combinations(row_list,2)))
count.most_common()[:-10-1:-1]   
```

## Finding out the number of movies done by each of the main cast actor for Christmas Movies released from 2017-22.

```python
#number of films done by each of the main cast actors
newdf = df.groupby("cast1")["cast1"].count()
newdf.sort_values(ascending=False)
```

## Finding out the most and least voted Christmas Movies from the ones that were released from 2017-2022.

```python
#removing the commas from votes and converting the values into int type
#printing the 5 most voted movie released from 2017-2022
df["votes"] = df["votes"].str.replace(',', '').astype(int)
df.sort_values("votes", ascending=False)
df.head()
```

```python
#printing the 5 least voted movie released from 2017-2022
df.sort_values("votes")
df.tail(5)
```

##Finding out the most and least used genre for Christmas Movies released from 2017-2022.

```python
#Cleaning the data so that 'Drama' and ' Drama' is being read as same
df["genre"] = df["genre"].str.replace(' ', '').astype(str)
df.sort_values("genre", ascending=False)
df.head()
```

```python
#10 most common genre appeared in the christmas movies from 2017-2022
from itertools import combinations
from collections import Counter

count = Counter()
for row in df["genre"]:
  row_list = row.split(',') #Creating the sublist and segrating by each comma
  count.update(Counter(combinations(row_list,1)))
count.most_common(5)
```

```python
#10 most uncommon genres appeared in the christmas movies from 2017-2022
from itertools import combinations
from collections import Counter

count = Counter()
for row in df["genre"]:
  row_list = row.split(',') #Creating the sublist and segrating by each comma
  count.update(Counter(combinations(row_list,1)))
count.most_common()[:-5-1:-1]  
```

## Finding out the longest and shortest Christmas Movies released from 2017-22.

```python
#converting the first 3 elements of time as a single integer and the sorting and printing the longest movies
df["time(in mins)"] = df["time"].str[0:3].astype(int)
df.sort_values("time(in mins)", ascending=False)
```

```python
#5 shortest movies
df.sort_values("time(in mins)", ascending=True)
df.tail()
```

## Finding all the Christmas Movies that start with C, have a rating 6.0 + and are of Drama genre.

```python
#Filtering the movies starting with C
import re
newdf = (df.loc[df["movie_title"].str.contains('^c[a-z]*', flags=re.I, regex=True)])
newdf
```

```python
#With the filtered set of movies that start with C, further filtering the ones that have a rating of 8.0 +
newdf["stars"].astype(float)
newdf = newdf[newdf["stars"] > 7.0]
newdf
```

```python
#Now after two filtering finally filtering the movies that are of genre 'Drama'
newdf.loc[newdf["genre"].str.contains("Drama")]
```

## Finding out the director who has worked most and least on Christmas Movies released from 2017-22.

```python
#number of films done by each of the directors
newdf = df.groupby("director_name")["director_name"].count()
newdf.sort_values(ascending=False)
```

##Finding out the per category/certificate count for the Christmas Movies released from 2017-2022.

```python
newdf = df.groupby("certificate")["certificate"].count()
newdf.sort_values(ascending=False)
```

## Finding out the exactly 7.1 rated Christmas Movies which were released from 2017-2022.

```python
df["stars"] = df["stars"].astype(str)
df.loc[df["stars"].str.contains("7.1")]
```

## Finding out the Christmas Movies that have 'cookie' in their description.

```python
df.loc[df["description"].str.contains('cookie[a-z]*', flags=re.I, regex=True)]
```

## Finding out the most common cast (i.e., the 4 people who have appeared together the most) for the Christmas Movies which were released from 2017-22.

```python
from itertools import combinations
from collections import Counter

count = Counter()
for row in df["all_actors"]:
  row_list = row.split(',') #Creating the sublist and segrating by each comma
  count.update(Counter(combinations(row_list,4)))
count.most_common(1) 
```
